Show a beautiful pie chart of top seven states that have been affected by corona virus 
          1. Top seven states where active cases have been reported maximum
          2. Top seven states where cured cases have been reported maximum
          3. Top seven states where death cases have been reported maximum


# Top Seven Active Case 

```python
# Top seven active states
data_active = df.sort_values(["active"]  , ascending = False)
data_active = data_active.iloc[:7, :]
data_active
```

```python
############## Pie chart for active cases########################################
explode = [0,0,0,0,0,0 ,0]
colors = ["orange","c","y","g","r" ,"grey" , "brown"]
textprops = {"Fontsize":15}
wedgeprops = {"linewidth": 4 , "width": 1 , }
plt.figure(figsize=(16,9))
fontdict = {"Fontsize":15}
plt.title("Top seven corrona affected states \n \n Total active cases in india  : {} \n Total active cases in these states {} \n Total percentage of active cases in these states :  {}%".format(sum(df["active"]) , sum(data_active["active"])  , round(sum(data_active["active"]) / sum(df["active"])  , 4)* 100 ), fontdict=fontdict)
plt.pie(data_active["active"], labels = data_active["state_name"],explode = explode,colors = colors, autopct= "%0.2f%%",radius = 1,textprops = textprops,
       pctdistance = 0.6,labeldistance = 1.1 ,wedgeprops = wedgeprops , rotatelabels=False )
plt.legend(loc = 2)
```

Top seven states in Active case 

(No Name, Kerala, Maharastra, Mizoram, Karnataka, Tamil Nadu, Uttar Pradesh )

# Top Seven Cured Case 

```python
# Top seven active states
data_active = df.sort_values(["cured"]  , ascending = False)
data_active = data_active.iloc[:7, :]
data_active
```

```python
############## Pie chart for active cases########################################
explode = [0,0,0,0,0,0 ,0]
colors = ["y","r","orange","g","c" ,"grey" , "brown"]
textprops = {"Fontsize":15}
wedgeprops = {"linewidth": 4 , "width": 1 , }
plt.figure(figsize=(16,9))
fontdict = {"Fontsize":15}
plt.title("Top seven corrona affected states \n \n Total active cases in india  : {} \n Total active cases in these states {} \n Total percentage of active cases in these states :  {}%".format(sum(df["cured"]) , sum(data_active["cured"])  , round(sum(data_active["cured"]) / sum(df["cured"])  , 4)* 100 ), fontdict=fontdict)
plt.pie(data_active["cured"], labels = data_active["state_name"],explode = explode,colors = colors, autopct= "%0.2f%%",radius = 1,textprops = textprops,
       pctdistance = 0.6,labeldistance = 1.1 ,wedgeprops = wedgeprops , rotatelabels=False )
plt.legend(loc = 2)
```
